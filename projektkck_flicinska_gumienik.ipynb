## Wykrywanie mrugnięć

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import time
```

```python
df = pd.read_csv("moje_dane_bizon.txt",names = ['signal','time'])
print(df)
```

```python
t=df[df.signal !='start']['time'].astype(str)
t_sekundy=[]
for i in t:
    h,m,s = i.split(":")
    result=int(h)*3600 +int(m)*60 + float(s)
    t_sekundy.append(result)
t_sekundy
```

```python
signal = df[df.signal != 'start']['signal'].astype(float) 
#wycięcie "start", zmiana zmiennych sygnału na float
```

```python
#sygnał
plt.plot(t_sekundy, signal)
plt.xlabel('Czas [s]')
plt.ylabel('Apmlituda [uV]')
```

```python
mrugniecia=[]
liczba = -40000
for value in df['signal']:
    if value == "start":
        pass
    elif float(value) < liczba:
        mrugniecia.append(1)
    else:
        mrugniecia.append(0)
print(mrugniecia)
```

```python
plt.plot(t_sekundy, mrugniecia)
plt.xlabel('Czas [s]')
```

```python
#zmiana sekund z czasu rzeczywistego na rozpoczynające się od 1
sekundy2 =[]
for sekunda in t_sekundy:
    sekundy2.append(sekunda - 44083.230518)
print(sekundy2)
```

```python
plt.plot(sekundy2, mrugniecia)
plt.xlabel('Czas [s]')
plt.ylabel('Występowanie mrugnięcia')
```

```python
#sygnał
plt.plot(sekundy2, signal)
plt.xlabel('Czas [s]')
plt.ylabel('Apmlituda [uV]')
```

```python
# 10-sekundowy fragment - litera B (sekunda 2)
t2 = np.linspace(0,10,2000)

plt.plot(t2, signal[2000:4000])
plt.xlabel('Czas [s]')
plt.ylabel('Apmlituda [uV]')
```

```python
#tworzenie słownika, który przyporządkowuje do czasu rzeczywistego fakt czy było mrugnięcie czy nie
slownik = dict(zip(sekundy2, mrugniecia))
#print(slownik)
slowo=[]
#następnym krokiem, jest usunięcie ze słownika pozycji, których definicja wynosi "0". Wtedy w słowniku powinny pozostać tylko pary czas:mrugnięcie.

for key, value in list(slownik.items()):
    if (value == 0):
        del(slownik[key])
print(slownik)
```

```python
#pętla do odczytywania słowa
ostateczne_slowo = []
for key, value in list(slownik.items()):
    if key > 2 and key < 3:
        if "B" not in ostateczne_slowo:
            ostateczne_slowo.append("B")
    elif key > 13 and key < 14:
        if "I" not in ostateczne_slowo:
            ostateczne_slowo.append("I")
    elif key > 31 and key < 32:
        if "Z" not in ostateczne_slowo:
            ostateczne_slowo.append("Z")
    elif key > 57 and key < 58:
        if "O" not in ostateczne_slowo:
            ostateczne_slowo.append("O")
    elif key > 83 and key < 84:
        if "N" not in ostateczne_slowo:
            ostateczne_slowo.append("N")
        
print(ostateczne_slowo)
```
